# Cause of Death in Germany | Data Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set()
%matplotlib inline
```

## Import Data

```python
df = pd.read_csv('../data/cause_of_death_germany_until_2018.csv', index_col=0)
```

```python
df.head()
```

## Plot timelines

**Death by year**

```python
ax = df.groupby(['year']).sum().plot()
ax.set_ylim(bottom=0, top=df.groupby(['year']).sum().max()['count']*1.2)
```

```python
data = df.groupby(['year', 'cause'], as_index=False).sum()
g = sns.FacetGrid(data, col="cause", col_wrap=6, height=2.5)
g = g.map(plt.plot, "year", "count", marker=".")
```

## Identify causes with the biggest change over the years (up or down)

**Filter the dataset for min and max year entries only**

```python
max_year = df['year'].unique()[0]
min_year = df['year'].unique()[-1]
```

```python
df_trend = df[df['year'].isin([max_year, min_year])]
```

**Remove gender and age**

```python
df_trend = df_trend.groupby(['year', 'cause'], as_index=False).sum()
```

```python
df_trend.head()
```

**Pivot df and add change over the years as percentage**

```python
df_trend = df_trend.pivot(index='cause', columns='year')
```

```python
df_trend.columns = [col[1] for col in df_trend.columns]
```

```python
df_trend.head()
```

```python
df_trend['change'] = df_trend[min_year] - df_trend[max_year]
df_trend['trend'] = df_trend['change'].apply(lambda x: 'up' if x > 0 else 'down')
df_trend['change_abs'] = df_trend['change'].apply(lambda x: np.abs(x))
df_trend['change_percent'] = df_trend['change']/df_trend[min_year]*100
```

```python
df_trend.head(2)
```

**Plot top 5 in change (percentage)**

```python
df_trend.sort_values('change_percent', ascending=False)[:5]['change_percent'].plot(kind='barh')
```

**Plot top 5 in upwards change**

```python
df_trend[df_trend['trend']=='up'].sort_values('change_percent', ascending=False)[:5]['change_percent'].plot(kind='barh')
```

**Plot top 5 in downwards change**

```python
df_trend[df_trend['trend']=='down'].sort_values('change_percent', ascending=False)[:5]['change_percent'].plot(kind='barh')
```

## Show the latest top causes 

```python
top_causes = df[df['year']==max_year].groupby(['cause'], as_index=False).sum().sort_values('count', ascending=False)['cause'][:20]
```

```python
df_total = df[(df['year']==max_year) &(df['cause'].isin(top_causes))].groupby(['cause', 'gender'], as_index=False).sum().drop(columns='year').sort_values('count', ascending=False)
```

```python
plt.figure(figsize=(10,20))
ax = sns.barplot(data=df_total, x='count', y='cause', hue='gender')
```
